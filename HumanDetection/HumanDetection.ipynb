### Convert Video to Images & Predict

```python
import cv2
import os
%cd /content/mmdetection
video = cv2.VideoCapture('1_00_01.mp4')
currentFrame = 0
directory1 = './1_00_01/'
if not os.path.exists(directory1):
    os.makedirs(directory1)
```

```python
while True:
    success,image = video.read()
    if not success:
        break
    filename = os.path.join(directory1, 'frame{:0>3d}.jpg'.format(currentFrame))
    cv2.imwrite(filename, image)
    os.system("python demo/image_demo.py "+filename+" /content/mmdetection/rtmdet_tiny_8xb32-300e_coco.py --weights /content/mmdetection/rtmdet_tiny_8xb32-300e_coco_20220902_112414-78e30dcc.pth --device cuda")
    print('Output Frame {} to file {}'.format(currentFrame, filename))
    currentFrame = currentFrame + 1

video.release()
```

```python
filename = '1_00_01_out.mp4'
fourcc = cv2.VideoWriter_fourcc(*'MP4V')
fps = 25


directory4 = './outputs/vis/'
images = [f for f in os.listdir(directory4)]
img = cv2.imread(os.path.join(directory4,images[0]))

output_width = img.shape[1]
output_height = img.shape[0]
images.sort()
writer = cv2.VideoWriter(filename, fourcc, fps, (output_width,output_height))
for f in images:
    fullpath = os.path.join(directory4,f)
    image = cv2.imread(fullpath)
    writer.write(image)
writer.release()
```
