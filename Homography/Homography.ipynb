```python
import numpy as np
import cv2

court2D_ics= np.array([[392, 378], [888, 378], [285, 677], [999, 677]])
court2D_wcs= np.array([[0, 0], [610, 0], [0, 1340], [610, 1340]])

# find homography matrix
H, status= cv2.findHomography(court2D_ics, court2D_wcs)

p_court2D_ics= [392, 378, 1]

# get x', y', w'
p_court2D_wcs= H@p_court2D_ics

# get x= x'/w', y= y'/w'
p_court2D_wcs= p_court2D_wcs/p_court2D_wcs[2]

print(p_court2D_wcs)

```
