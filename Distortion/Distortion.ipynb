# Import library and setting parameter

```python
import cv2
import numpy as np

# distortion parameter
dist = np.array([[-0.5174073649660939, 0.3434369569637, 0.00017737728144079067, 0.0012798953431086466, -0.14252643754041577]])
# intrinsic matrix
mtx = np.array([[1235.057507552043, 0.0, 762.6318057740573], [0.0, 1235.887521219294, 554.1490296341883], [0.0, 0.0, 1.0]])
```

```python
img = cv2.imread('distortion.png')

width = img.shape[0]
height = img.shape[1]

newcameramtx1, roi= cv2.getOptimalNewCameraMatrix(mtx, dist, (width,height), 1, (width,height))
undistorted_frame1 = cv2.undistort(
    img, mtx, dist, None, newcameramtx1)
cv2.imshow("undistorted %s" % (undistorted_frame1.shape,), undistorted_frame1)
cv2.waitKey(10) 
cv2.imwrite('undistortion.png', undistorted_frame1)

newcameramtx2, roi= cv2.getOptimalNewCameraMatrix(mtx, dist, (width,height), 0, (width,height))
undistorted_frame2 = cv2.undistort(
    img, mtx, dist, None, newcameramtx2)
cv2.imshow("undistorted %s" % (undistorted_frame2.shape,), undistorted_frame2)
cv2.waitKey(10) 
cv2.imwrite('undistortion_filter.png', undistorted_frame2)
```
